```python
import pandas as pd
df1 = pd.read_csv('bengaluru_house_prices.csv')
df1
```

```python
df1.isna().sum()
```

```python
df2 = df1.drop(['area_type', 'availability', 'society', 'balcony'], axis='columns')
df2
```

```python
df3 = df2.dropna()
df3
```

```python
df3['size'].unique()
```

```python
df4 = df3.copy()
df4['bhk'] = df3['size'].apply(lambda x: float(x.split(' ')[0]))
df4.drop(['size'], axis = 'columns', inplace=True)
df4
```

```python
df4.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df4[~df4.total_sqft.apply(is_float)].total_sqft
```

```python
def convert_to_float(x):
    try:
        return float(x)
    except:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1]))/2
        return -1
```

```python
df5 = df4[~(df4.total_sqft.apply(convert_to_float) < 0)]
df5
```

```python
df5.loc[:,'total_sqft'] = df5.total_sqft.apply(convert_to_float)
df5.total_sqft.unique()
```

```python
df6 = df5.copy()
df6['price_per_sqft'] = (df5.price * 100000 / df5.total_sqft)
df6
```

```python
df7 = df6[~(df6.total_sqft/df6.bhk<300)]
df7
```

```python
df8 = df7[(df7.bath - df7.bhk <3)] 
df8
```

```python
print(df8['price_per_sqft'].min(), df8['price_per_sqft'].mean(), df8['price_per_sqft'].std(), df8['price_per_sqft'].max())
```

```python
import numpy as np
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df9 = remove_pps_outliers(df8)
df9
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline 
```

```python
def scatter_plot(df, location):
    df = df[df.location==location]
    bhk2 = df[df.bhk==2]
    bhk3 = df[df.bhk==3]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    # plt.scatter(bhk1.total_sqft, bhk1.price, color='green', marker='+')
    plt.scatter(bhk2.total_sqft, bhk2.price, color='red', marker='+', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, color='blue', marker='+', label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
```

```python
scatter_plot(df9, 'Rajaji Nagar')
```

Remove outlier if in a location, price of n bhk is less than mean of price of (n-1) bhk

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df10 = remove_bhk_outliers(df9)
df10
```

```python
df10.loc[:,'location'] = df10['location'].apply(lambda x: x.strip())
df10
```

```python
len(location_stats[location_stats <= 10])
```

```python
df11 = df10.copy()
df11['location'] = df10.location.apply(lambda x: 'other' if x in location_stats[location_stats <= 10] else x)
df11.location
```

```python
df12 = df10.copy()
dummies = pd.get_dummies(df11.location)
df12 = pd.concat([df12, dummies], axis='columns')
df12.drop(['location'], inplace=True, axis='columns')
df12
```
